# Histogram equalization
- https://en.wikipedia.org/wiki/Histogram_eqaulization
 - uesd to equalization of the color
 - mostly used in the video editing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
gorilla= cv2.imread("C:/Artificial Intelligence/2_Computer_Vision/1_image_processing/gorilla.jpg",0)

'''
Here's a brief overview of the different options for this arguments.:
0: Read the image in the grayscale mode
1: read the image in the color mode(default), resulting in 3 channel immage
-1: read the image with the alpha channel , it it exists , resulting in 4 channel image

In your case using 0 will load gorilla.jpg as grayscale image
'''
```

```python
def display(img, cmap=None):
    fig=plt.figure(figsize=(10,8))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
display(gorilla, cmap='gray')
```

```python
hist_values=cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0,256])
'''
['gorilla']:
this is the image (or list of image ) for which you want to calculate the 

channels:[0]:

 This specifies the index of the channel you want to compute
 the histogram for. Since the image is grayscale, it has only one channels
 which is indexed as 0.
 For color images, you could specify [0] for the blue channel, [1] for the green channel
 and [2] for the red channel
 
 mask=None:
 

'''
```

```python
plt.plot(hist_values)
#in this value on;y the gray value garph is present
```

```python
#Equallization
```

```python
eq_gorilla=cv2.equalizeHist(gorilla)
```

```python
display(eq_gorilla, cmap='gray')
```

```python
hist_values=cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_values)
```
