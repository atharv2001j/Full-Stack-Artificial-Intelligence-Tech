# Drawing the Images

```python

import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
blank_img= np.zeros(shape=(512, 512, 3), dtype= np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

# Shape

# 1. Rectangle
- img. Image
- pt1 Vertex of the rectangle
-pt2 vertex of the rectangle opposite to pt1
- color Rectangle color or brightness (grayscale image)
- line type type of the line See # line type

```python
#pt1= top left
#pt2= bottom right

cv2.rectangle(blank_img, pt1=(384, 0), pt2=(510, 128), color= (0, 255, 0), thickness= 5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200, 200), pt2=(300, 300), color= (0, 0, 255), thickness= 5)
plt.imshow(blank_img)
```

# 2. Circle
- center
- radius

```python
cv2.circle(blank_img, center=(100, 100), radius=50, color= (255, 255, 0), thickness= 5)
plt.imshow(blank_img)
```

# filled IN Circle

```python
cv2.circle(blank_img, center=(400, 400), radius=50, color= (255, 255, 0), thickness= -1)
plt.imshow(blank_img)
```

# 3. lines
- pt1 
- pt2

```python
cv2.line(blank_img, pt1=(0,0), pt2=(511, 511), color= (102, 255, 255), thickness= 5)
plt.imshow(blank_img)
```

# 4. TEXT
- text 
- org
- fontScale
- lineType

```python
font = cv2. FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello', org=(10,500), fontFace=font, fontScale= 4 , color=(255, 255, 255), thickness=2, lineType= cv2.LINE_AA)
plt.imshow(blank_img)
```

pts=vertices.reshape((-1,1,2))

vertices.reshape((-1,1,2)) : reshape is a method that change the shape of the array without
