# Resize image By Ratio

```python
import cv2

#assume img_rgb is already loaded with an image
w_ratio=0.5 #Scaling factor for width
h_ratio= 0.5  #scaling factor for width

#resize the image
new_img = cv2.resize(img_rgb,(0,0) , fx =w_ratio , fy=h_ratio)


'''
cv2.resize()
This is a function from the opencv libraray that resize an image to a specific size

img_rgb:
This is the input image that you want to resize. IT is assumed to be in RGB Format.

(0,0):
This is the desired size of the output image
Setting both dimension to 0 means that the size willbe determined
by the scaling factors provided.

'''
```

```python
plt.imshow(new_img)
```

# Flipping Images

```python
%matplotlib inline
```

```python
#along centarl x axis
new_img= cv2.flip(new_img, 0)
plt.imshow(new_img)

'''
 This is the flip code that specifiec how the image should be flipped:
     0: mean s flipping around the x_axis (vertical flip)
     1: means flipping around the y_axis (horizontal flip)
     -1:  means flipping around the both axis (both vertical and horizontal flip)

'''
```

```python
#ALONG CENTRA Y AXIS
new_img= cv2.flip(new_img, 1)
plt.imshow(new_img)
```

```python
#Along both axis
new_img= cv2.flip(new_img, -1)
plt.imshow(new_img)
```
