```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
#generate the some random data
#to simulate the data
np.random.seed(42)
X=np.random.rand(100,10)
y=np.random.rand(100,1)
```

```python
#define the simple model
#input_dim means the number of the feature for the model in X
def create_model(loss):
    model=Sequential([
        Dense(64, input_dim=10, activation='relu'),
        Dense(64, activation='relu'),
        Dense(1, activation='linear')
    ])
    model.compile(optimizer=Adam(learning_rate=0.01), loss=loss, metrics=['mae'])
    return model
```

```python
#list of the loss function to demonstrate
loss_function={
        'Mean Squared Error':'mean_squared_error',
        'Mean Absolute Error':'mean_absolute_error',
        'Mean Squared Logarithmic error': 'mean_squared_logarithmic_error',
}

```

```python
#Traning the model with different loss function
history={}
for name, loss in loss_function.items():
    print(f"traning with{name} loss function")
    model=create_model(loss)
    history[name]=model.fit(X, y, epochs=10, batch_size= 10, verbose=0)
    print(f" Finished traning with {name} loss function \n" )
```

```python
#ploating the result
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
for name in loss_function.keys():
    plt.plot(history[name].history['loss'], label=f'{name} Loss')

plt.title('Loss per epoach for different loss function')
plt.xlabel('Epoach')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
