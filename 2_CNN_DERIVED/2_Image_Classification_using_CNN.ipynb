# Image manipulation
Its usually a good ideas to manipulate the image with rotate , resizing  and scaling so the model becomes more robust to different 
images that our dataset does not have. We can use the ImageDataGenerator to do automatically for us

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen=ImageDataGenerator(rotation_range=30, #rotate the image 30 degeers
                            width_shift_range=0.1, # shift the pic width by a max of 1st
                            height_shift_range=0.1, # shift the pic height by max of 1st
                            rescale=1/255, #rescale the image by normalizing it
                            shear_range=0.2, #shear the means cutting away part of the image
                            zoom_range=0.2, # Zoom in the 20 % max
                            horizontal_flip=True, # allo horizontal flipping
                            fill_mode='nearest' # Fill in missing pixel with nearset
                            )
```

# Resizing Images
Lets have keras resize all the images to 150 pixels by 150 pixels once they have been manipulated

```python
#width, height, channels
image_shape=(150,150,3)
```

# creating the model

```python
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
model= Sequential()

model.add(Conv2D(filters=32, Kernal_size=(3,3), input_shape=(150,150,3), activation='relu',))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64, Kernal_size=(3,3), input_shape=(150,150,3), activation='relu',))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64, Kernal_size=(3,3), input_shape=(150,150,3), activation='relu',))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
```

```python
model.add(Dense(128))
model.add(Activation('relu'))

#dropout help to reduce overfitting by randomly turing the neurons of during the traning
#here we say randomly turn off 50% of neurons

model.add(Dropout(0.5))

#last layers, remeber it binary 0=cat, 1=Dogs
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
